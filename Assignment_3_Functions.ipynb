**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

->
1.   Map():

    * The map function applies a given function to all items in an iterable (like a list).
    *   returns an iterator with the results.
    * Ex.
    num = [1, 2, 3, 4]

      s = map(lambda x: x**2, num)

      print(list(s))  # [1, 4, 9, 16]


2.  reduce()

    *  The reduce function from the functools module applies a given function cumulatively to the items of a sequence.
    *   from left to right, to reduce the sequence to a single value.
    *   Ex:
    
      from functools import reduce

      nums = [1, 2, 3, 4, 5]

      total = reduce(lambda x, y: x + y, nums)

      print(total)  # 15

2.  filter()

    *  The filter function constructs an iterator from elements of an iterable for which a specified function returns True.

    *   Used for filtering elements from an iterable based on a condition defined by the function.
    *   Ex:
    
      nums = [1, 2, 3, 4, 5]

      even = filter(lambda x: x % 2 == 0, nums)

      print(list(even))  // [2, 4]

```python
# 5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.

class Fib_Iterator:
    def __init__(self, total_terms):
        self.total_terms = total_terms
        self.generated = 0
        self.first = 0
        self.second = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.generated >= self.total_terms:
            raise StopIteration
        if self.generated == 0:
            self.generated += 1
            return self.first
        elif self.generated == 1:
            self.generated += 1
            return self.second
        else:
            next_value = self.first + self.second
            self.first, self.second = self.second, next_value
            self.generated += 1
            return next_value

# Example usage
fib_sequence = Fib_Iterator(5)
for number in fib_sequence:
    print(number)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power(ip_exponent):
    for i in range(ip_exponent + 1):
        yield 2 ** i

e=power(5)
for i in range(5):
  print(next(e))
```

```python
#  10.Create a Python program that uses filter() to remove all the vowels from a given string.
vowels = "aeiouAEIOU"
string = input("Enter the string: ")

result = filter(lambda x: x not in vowels, string)

without_vowels = ''.join(result)

print("With Vowels:", string)
print("Without Vowels:", without_vowels)
```
