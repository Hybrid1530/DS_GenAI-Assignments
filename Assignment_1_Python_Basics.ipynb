```python
#5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently

# List
my_list = [10, 20, 30,50]
my_list[1] = 99
print("Modified list:", my_list)

# Tuple (immutable, can't change)
my_tuple = (10, 20, 30)
try:
    my_tuple[1] = 99
except TypeError as err:
    print(f"Error when modifying tuple: {err}")
```

```python
# 6.Write a function to demonstrate the behavior of mutable and immutable arguments.
def modify_args(num, my_list):
    num += 10             # Immutable: integer
    my_list.append(100)   # Mutable: list
    print("Inside function - num:", num)
    print("Inside function - list:", my_list)

x = 5                    # Immutable object
lst = [1, 2, 3]          # Mutable object

modify_args(x, lst)

print("Outside function - x:", x)        # Unchanged
print("Outside function - lst:", lst)    # Changed
```

```python

#7. Write a program that performs basic arithmetic operations on two user-input numbers.
print("Basic Calculator")
value1=int(input("Enter the 1st number: "))
value2=int(input("Enter the 2nd number: "))
operation=input(" Enter the opertion to perform: " '"+","-","*","/" ')

if(operation=="+"):
  print(value1 + value2)

elif(operation=="-"):
  print(value1 - value2)

elif(operation=="*"):
    print(value1 * value2)

elif(operation=="/"):
  print(float(value1 / value2))
else:
  print("Wrong input !!")


```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num=int(input("Enter the number to Calculate it's Factorial: "))

factorial = 1
i = 1
if(num>0):
  while i <= num:
    factorial *= i
    i += 1
  print("Factorial of the  number is", factorial)
else:
  print("Enter the factorial number again")

```
