**2.How does broadcasting work in NumPy?**

->

- Broadcasting is a mechanism in NumPy that allows performing operations on arrays of different shapes under certain conditions.
- When operands have different shapes, NumPy expands the smaller array to match the larger one in a specific way to perform element-wise operations.

ex:

    import numpy as np
    a = np.array([1, 2, 3])
    b = 2
    print(a + b)  //Output: [3 4 5]
                  //b is broadcast to [2, 2, 2]

**4.Explain the use of the groupby() method in Pandas?**

->The groupby() method is used to group data based on one or more keys (columns), then apply aggregation functions like sum(), mean(), etc.

Ex:

    import pandas as pd
    data = {
    'Department': ['HR', 'HR', 'IT', 'IT'],
    'Salary': [3000, 4000, 5000, 6000]
    }
    df = pd.DataFrame(data)
    grouped = df.groupby('Department')['Salary'].mean()
    print(grouped)

output will be like this,

    Department
    HR    3500.0
    IT    5500.0

**17.Explain the difference between apply() and map() in Pandas?**

->map()
- Used only on Series.
- Applies a function element-wise to a Pandas Series (single column).
- Can take a function, dictionary, or a Series for mapping.

Example:

    df['Age'].map(lambda x: x + 1)

apply()
- Used on Series or DataFrame.
- More versatile.
- Can apply a function to Each element of a Series, or Each row or column of a DataFrame.

Example:

    df['Age'].apply(lambda x: x + 1)

**21.Why is NumPy’s array slicing faster than Python’s list slicing?**

->
1. Fixed Data Type: NumPy arrays store data in a contiguous block of memory with a fixed type (e.g., float64), while Python lists can hold mixed types and are stored as references.
2. Vectorized Operations: NumPy uses compiled C code under the hood for operations like slicing, avoiding the overhead of Python loops.
3. View Instead of Copy: Slicing a NumPy array returns a view (not a full copy), which is much faster and memory-efficient. Python lists create a new list during slicing.

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")


corr_matrix = df.corr(numeric_only=True)


plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
```
