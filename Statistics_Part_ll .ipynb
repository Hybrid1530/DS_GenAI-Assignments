**1.What is hypothesis testing in statistics?**

->Hypothesis testing is a statistical method used to make inferences or decisions about population parameters based on sample data. It helps determine whether the observed data provides enough evidence to reject a certain assumption (the null hypothesis).

**2.What is the null hypothesis, and how does it differ from the alternative hypothesis?**

->Null hypothesis (H₀): A default assumption that there is no effect or difference.

Alternative hypothesis (H₁ or Ha): States that there is an effect or difference.

Example:
1. H₀: μ = 100 (the mean is 100)
2. H₁: μ != 100 (the mean is not 100)

**3.What is the significance level in hypothesis testing, and why is it important?**

->Significance level (α): The probability of rejecting the null hypothesis when it is actually true (Type I error).

Common values: 0.05, 0.01, or 0.10.

It sets the threshold for deciding whether an observed result is statistically significant.

**4.What does a P-value represent in hypothesis testing?**

->The P-value is the probability of obtaining a result as extreme as or more extreme than the observed result, assuming that the null hypothesis is true.

**5.How do you interpret the P-value in hypothesis testing?**

If P-value ≤ Significance level reject Null hypothesis  → result is statistically significant.

If P-value > Significance level, fail to reject Null hypothesis → not statistically significant.

E.g., P = 0.03 with α = 0.05 ⇒ reject Null hypothesis .

**6.What are Type I and Type II errors in hypothesis testing?**

->
1. Type I error (False Positive): Rejecting H₀ when it is true (controlled by significance level).

2. Type II error (False Negative): Failing to reject H₀ when H₁ is true (denoted by β).

Power = 1 − β (probability of correctly rejecting H₀).

**7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

->
* One-tailed test: Tests for deviation in one direction (e.g., H₁: μ > 100).
* Two-tailed test: Tests for deviation in either direction (e.g., H₁: μ ≠ 100).
Choice depends on the research question.

**8.What is the Z-test, and when is it used in hypothesis testing?**

->A Z-test is used to test hypotheses about population means (or proportions) when:
* The population standard deviation is known
* Sample size is large (n ≥ 30)
* Used in one-sample or two-sample comparisons.

**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**

->
Z = (X̄ − μ) / (σ / √n)

Z-score is difference of the sample mean(X̄) and the population mean(μ) divided by the result of population standard deviation(σ) divided by root of sample size(n).
This score is then check with the Z table to get the corresponding value(if needed).

The Z-score tells you how many standard deviations the sample mean is from the population mean.

Example: A 95% CI means we are 95% confident that the interval contains the true mean.
CI = X̄ ± (critical value × standard error)

**15.What is the margin of error, and how does it affect the confidence interval?**

* ME = critical value × standard error
* Larger ME = wider CI = less precision
* Decreases with larger sample size or lower confidence level.

```python
#1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy import stats


def z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    # Step 1: Calculate sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Step 2: Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Step 3: Compute P-value (two-tailed)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Step 4: Interpretation
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Score: {z_score:.3f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (H₀).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (H₀).")



# Example usage
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=50)  # Simulated sample
population_mean = 100
population_std = 15

z_test(sample, population_mean, population_std)
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def two_tailed_z_test_with_plot(sample, pop_mean, pop_std, alpha=0.05):
    # Step 1: Calculate sample statistics
    sample_mean = np.mean(sample)
    n = len(sample)

    # Step 2: Perform two-tailed Z-test
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed

    # Step 3: Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Score: {z_score:.3f}")
    print(f"P-Value: {p_value:.4f}")
    if p_value < alpha:
        print("Conclusion: Reject H₀ (Statistically significant difference)")
    else:
        print("Conclusion: Fail to reject H₀ (No significant difference)")

    # Step 4: Visualization of the decision region
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    z_crit = stats.norm.ppf(1 - alpha / 2)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Rejection regions
    plt.fill_between(x, y, where=(x <= -z_crit) | (x >= z_crit),
                     color='red', alpha=0.3, label='Rejection Region (α = 0.05)')

    # Plot Z-score
    plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')

    # Labels and styling
    plt.title('Two-Tailed Z-Test with Decision Regions')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: Simulate a sample
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=15, size=50)

# Known population parameters
population_mean = 100
population_std = 15

# Run the two-tailed Z-test and plot
two_tailed_z_test_with_plot(sample_data, population_mean, population_std)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu_0=100, mu_a=105, sigma=10, n=30, alpha=0.05):
    # Step 1: Calculate standard error
    se = sigma / np.sqrt(n)

    # Step 2: Critical Z value and critical X values for two-tailed test
    z_crit = norm.ppf(1 - alpha / 2)
    x_lower = mu_0 - z_crit * se
    x_upper = mu_0 + z_crit * se

    # Step 3: Range of x-values for plotting
    x = np.linspace(mu_0 - 4*se, mu_a + 4*se, 1000)

    # Step 4: Normal distributions under H₀ and H₁
    y_h0 = norm.pdf(x, mu_0, se)
    y_h1 = norm.pdf(x, mu_a, se)

    # Step 5: Plot distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_h0, label='H₀: μ = {}'.format(mu_0), color='blue')
    plt.plot(x, y_h1, label='H₁: μ = {}'.format(mu_a), color='green')

    # Step 6: Shade Type I Error (α) region under H₀
    plt.fill_between(x, y_h0, where=(x < x_lower) | (x > x_upper),
                     color='red', alpha=0.3, label='Type I Error (α)')

    # Step 7: Shade Type II Error (β) region under H₁
    plt.fill_between(x, y_h1, where=(x > x_lower) & (x < x_upper),
                     color='orange', alpha=0.3, label='Type II Error (β)')

    # Plot details
    plt.axvline(x_lower, color='black', linestyle='--')
    plt.axvline(x_upper, color='black', linestyle='--')
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Test Statistic (X̄)")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the function
plot_type1_type2_errors()
```

```python
#6.Write a Python program to perform an independent T-test and interpret the results

import numpy as np
from scipy import stats

def independent_t_test(group1, group2, alpha=0.05):
    # Step 1: Perform independent t-test (assumes equal variances by default)
    t_stat, p_value = stats.ttest_ind(group1, group2)

    # Step 2: Print results
    print(f"Group 1 Mean: {np.mean(group1):.2f}")
    print(f"Group 2 Mean: {np.mean(group2):.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_value:.4f}")

    # Step 3: Interpretation
    if p_value < alpha:
        print("Conclusion: Reject H₀ (Significant difference between groups)")
    else:
        print("Conclusion: Fail to reject H₀ (No significant difference between groups)")

# Simulate sample data for two independent groups
np.random.seed(42)
group1 = np.random.normal(loc=100, scale=10, size=30)  # Mean = 100
group2 = np.random.normal(loc=105, scale=10, size=30)  # Mean = 105

# Run the t-test
independent_t_test(group1, group2)
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def paired_t_test_with_plot(before, after, alpha=0.05):
    # Step 1: Perform paired t-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Step 2: Print results
    print(f"Mean (Before): {np.mean(before):.2f}")
    print(f"Mean (After): {np.mean(after):.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_value:.4f}")

    # Step 3: Interpretation
    if p_value < alpha:
        print("Conclusion: Reject H₀ (Significant difference between before and after)")
    else:
        print("Conclusion: Fail to reject H₀ (No significant difference between before and after)")

    # Step 4: Visualization
    plt.figure(figsize=(8, 5))
    for i in range(len(before)):
        plt.plot([0, 1], [before[i], after[i]], marker='o', color='red', alpha=0.7)

    plt.xticks([0, 1], ['Before', 'After'])
    plt.ylabel("Values")
    plt.title("Paired Sample Comparison (Before vs After)")
    plt.grid(True)
    plt.show()

# Simulate paired data (e.g., test scores before and after a training)
np.random.seed(42)
before_scores = np.random.normal(loc=100, scale=8, size=30)
after_scores = before_scores + np.random.normal(loc=2, scale=5, size=30)  # slight improvement

# Run the paired t-test and plot
paired_t_test_with_plot(before_scores, after_scores)
```

```python
#8.Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
from scipy import stats

def compare_z_and_t_test(pop_mean=100, pop_std=15, sample_size=25, alpha=0.05):
    # Step 1: Simulate sample data
    np.random.seed(42)
    sample = np.random.normal(loc=102, scale=pop_std, size=sample_size)

    # Step 2: Z-test (population std known)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    z_p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Step 3: T-test (population std unknown)
    t_stat, t_p_value = stats.ttest_1samp(sample, pop_mean)

    # Step 4: Display results
    print("Sample Mean:", round(sample_mean, 2))
    print("Sample Size:", sample_size)
    print("-" * 45)
    print("Z-Test (σ known):")
    print(f"  Z-score   = {z_score:.3f}")
    print(f"  P-value   = {z_p_value:.4f}")
    print(f"  Conclusion: {'Reject H₀' if z_p_value < alpha else 'Fail to reject H₀'}")
    print("-" * 45)
    print("T-Test (σ unknown):")
    print(f"  T-statistic = {t_stat:.3f}")
    print(f"  P-value     = {t_p_value:.4f}")
    print(f"  Conclusion: {'Reject H₀' if t_p_value < alpha else 'Fail to reject H₀'}")

# Run the comparison
compare_z_and_t_test()
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(sample, confidence=0.95):
    # Step 1: Sample statistics
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # use sample std deviation
    se = sample_std / np.sqrt(n)

    # Step 2: T critical value for the desired confidence level
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    # Step 3: Margin of Error
    margin_of_error = t_crit * se

    # Step 4: Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Step 5: Output
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Sample Size = {n}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print("\n Interpretation:")
    print(f"We are {int(confidence*100)}% confident that the true population mean lies within this interval.")

    print(f"lower_bound={round(lower_bound, 2)} \n upper_bound = {round(upper_bound, 2)}")
    return

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=10, size=30)
confidence_interval(sample_data, confidence=0.95)
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats

def calculate_margin_of_error(sample, confidence=0.95):
    # Step 1: Sample statistics
    n = len(sample)
    sample_std = np.std(sample, ddof=1)  # sample standard deviation
    se = sample_std / np.sqrt(n)         # standard error

    # Step 2: T critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    # Step 3: Margin of Error
    margin_of_error = t_crit * se

    # Step 4: Output
    print(f"Sample Size: {n}")
    print(f"Sample Std Dev: {sample_std:.2f}")
    print(f"Standard Error: {se:.2f}")
    print(f"T-critical value ({int(confidence*100)}%): {t_crit:.3f}")
    print(f"\n Margin of Error: {margin_of_error:.2f}")

    return round(margin_of_error, 2)

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=10, size=30)
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

"""
Problem Example: Diagnostic Test for a Rare Disease
Given:
P(Disease) = 1% → prior probability

P(No Disease) = 99%

P(Positive Test | Disease) = 0.95 → sensitivity (true positive rate)

P(Positive Test | No Disease) = 0.05 → false positive rate

Goal:
What is P(Disease | Positive Test) — i.e., the posterior probability someone has the disease given they tested positive?

"""
def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_notA):
    """
    Compute P(A | B) using Bayes' Theorem.

    Parameters:
    - prior_A: P(A) = Prior probability of hypothesis (e.g., has disease)
    - prob_B_given_A: P(B | A) = Likelihood (e.g., test positive given disease)
    - prob_B_given_notA: P(B | ¬A) = False positive rate (test positive without disease)

    Returns:
    - posterior: P(A | B) = Updated probability given evidence
    """
    prior_notA = 1 - prior_A

    # Total probability of B (evidence)
    prob_B = (prob_B_given_A * prior_A) + (prob_B_given_notA * prior_notA)

    # Apply Bayes' Theorem
    posterior = (prob_B_given_A * prior_A) / prob_B

    # Output
    print(f"P(A | B) = {posterior:.4f}")
    return posterior

# Example values
prior_disease = 0.01            # 1% have the disease
sensitivity = 0.95              # P(Positive | Disease)
false_positive_rate = 0.05      # P(Positive | No Disease)

# Run Bayes' Theorem
posterior_prob = bayes_theorem(prior_disease, sensitivity, false_positive_rate)
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python

import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table (e.g., Gender vs Purchase Decision)
data = {
    'Buy': [30, 20],      # [Males, Females]
    'Not Buy': [10, 40]
}
table = pd.DataFrame(data, index=['Male', 'Female'])

# Step 2: Perform the chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(table)

# Step 3: Print results
print("Contingency Table:")
print(table)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=table.index, columns=table.columns))

print(f"\nChi-square Statistic = {chi2_stat:.3f}")
print(f"Degrees of Freedom = {dof}")
print(f"P-value = {p_val:.4f}")

# Step 4: Interpretation
alpha = 0.05
if p_val < alpha:
    print(" Conclusion: Reject H₀ → Variables are dependent.")
else:
    print(" Conclusion: Fail to reject H₀ → No evidence of dependence.")
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected):
    # Perform Chi-square goodness-of-fit test
    chi_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

    # Output results
    print("Observed Frequencies:", observed)
    print("Expected Frequencies:", expected)
    print(f"\nChi-square Statistic = {chi_stat:.3f}")
    print(f"P-value = {p_val:.4f}")

    # Interpretation
    alpha = 0.05
    if p_val < alpha:
        print(" Conclusion: Reject H₀ → Observed data does NOT fit the expected distribution.")
    else:
        print(" Conclusion: Fail to reject H₀ → Observed data fits the expected distribution.")

# Example: Observed dice roll frequencies (not perfectly fair)
observed = np.array([12, 8, 11, 9, 10, 10])

# Expected: Fair dice (uniform distribution)
expected = np.array([10, 10, 10, 10, 10, 10])

# Run the test
goodness_of_fit_test(observed, expected)
```

```python
#16.Implement an F-test using Python to compare the variances of two random samples

import numpy as np
from scipy.stats import f

def f_test_variance(sample1, sample2, alpha=0.05):
    # Step 1: Sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Step 2: Calculate F-statistic (larger variance / smaller variance)
    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    # Step 3: Critical value for two-tailed test
    f_critical = f.ppf(1 - alpha / 2, dfn, dfd)
    f_lower = f.ppf(alpha / 2, dfn, dfd)

    # Step 4: Output
    print(f"Sample 1 Variance = {var1:.2f}")
    print(f"Sample 2 Variance = {var2:.2f}")
    print(f"F-statistic = {f_stat:.3f}")
    print(f"F-critical values (two-tailed): Lower = {f_lower:.3f}, Upper = {f_critical:.3f}")

    # Step 5: Conclusion
    if f_stat > f_critical or f_stat < f_lower:
        print(" Conclusion: Reject H₀ → Variances are significantly different.")
    else:
        print(" Conclusion: Fail to reject H₀ → No significant difference in variances.")

# Simulate two independent samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # std = 10
sample2 = np.random.normal(loc=50, scale=15, size=30)  # std = 15

# Run F-test
f_test_variance(sample1, sample2)
```

```python
#19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")

def check_anova_assumptions(data, group_col, value_col):
    """
    Check ANOVA assumptions: normality, independence (by design), and equal variance.

    Parameters:
    - data: DataFrame containing the data
    - group_col: column name indicating the group/category
    - value_col: column name indicating the measurement/score
    """

    groups = data[group_col].unique()

    print("Checking Normality (Shapiro-Wilk Test):")
    for g in groups:
        stat, p = shapiro(data[data[group_col] == g][value_col])
        print(f"  Group {g}: p-value = {p:.4f} → {'Pass ' if p > 0.05 else 'Fail '}")

    print("\n Checking Equal Variance (Levene's Test):")
    group_values = [data[data[group_col] == g][value_col] for g in groups]
    stat, p = levene(*group_values)
    print(f"  Levene’s test p-value = {p:.4f} → {'Pass ' if p > 0.05 else 'Fail '}")

    print("\n Q-Q Plots for Normality:")
    for g in groups:
        sm.qqplot(data[data[group_col] == g][value_col], line='s')
        plt.title(f'Q-Q Plot: Group {g}')
        plt.grid(True)
        plt.show()

    print(" Independence must be ensured by study design (e.g., random sampling).")

# Simulate data
np.random.seed(42)
group_A = np.random.normal(50, 5, 30)
group_B = np.random.normal(52, 5, 30)
group_C = np.random.normal(55, 5, 30)

df = pd.DataFrame({
    'Score': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A']*30 + ['B']*30 + ['C']*30
})

# Check assumptions
check_anova_assumptions(df, group_col='Group', value_col='Score')
```

```python
#20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Step 1: Simulate data
np.random.seed(42)
gender = ['Male', 'Female']
treatment = ['A', 'B']

data = []

for g in gender:
    for t in treatment:
        # Add random normal scores with different means for interaction
        mean = 50 if g == 'Male' and t == 'A' else \
               60 if g == 'Male' and t == 'B' else \
               55 if g == 'Female' and t == 'A' else \
               65
        scores = np.random.normal(loc=mean, scale=5, size=20)
        for score in scores:
            data.append({'Gender': g, 'Treatment': t, 'Score': score})

df = pd.DataFrame(data)

# Step 2: Perform two-way ANOVA
model = ols('Score ~ C(Gender) + C(Treatment) + C(Gender):C(Treatment)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\n🔍 Two-Way ANOVA Results:")
print(anova_table)

# Step 3: Visualization - Interaction Plot
plt.figure(figsize=(8, 6))
sns.pointplot(x="Treatment", y="Score", hue="Gender", data=df, dodge=True,
              markers=["o", "s"], capsize=0.1, errwidth=1.5, palette="Set2")
plt.title("Interaction Plot: Treatment vs Gender")
plt.grid(True)
plt.show()
```

```python
#24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
from scipy.stats import chi2

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=6, size=30)  # std = 6 → variance = 36

# Step 2: Sample statistics
sample_var = np.var(sample, ddof=1)  # unbiased estimator
n = len(sample)

# Step 3: Hypothesized population variance (σ₀²)
sigma0_sq = 25  # e.g., hypothesized std = 5

# Step 4: Chi-square test statistic
chi2_stat = (n - 1) * sample_var / sigma0_sq

# Step 5: Critical values for two-tailed test
alpha = 0.05
df = n - 1
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)

# Step 6: Output results
print("🔍 Chi-square Test for Population Variance")
print(f"Sample Variance     = {sample_var:.2f}")
print(f"Hypothesized Variance = {sigma0_sq}")
print(f"Chi-square Statistic = {chi2_stat:.2f}")
print(f"Degrees of Freedom   = {df}")
print(f"Chi-square Critical Values: [{chi2_lower:.2f}, {chi2_upper:.2f}]")

# Step 7: Interpretation
if chi2_stat < chi2_lower or chi2_stat > chi2_upper:
    print(" Conclusion: Reject H₀ → Variance is significantly different from hypothesized value.")
else:
    print(" Conclusion: Fail to reject H₀ → No significant difference in variance.")
```

```python
#26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the result

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Step 1: Simulate two datasets
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # variance ≈ 100
sample2 = np.random.normal(loc=50, scale=15, size=30)  # variance ≈ 225

# Step 2: Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Step 3: Compute F-statistic (larger variance / smaller variance)
if var1 > var2:
    f_stat = var1 / var2
    dfn, dfd = len(sample1) - 1, len(sample2) - 1
else:
    f_stat = var2 / var1
    dfn, dfd = len(sample2) - 1, len(sample1) - 1

# Step 4: Critical values for two-tailed test
alpha = 0.05
f_crit_low = f.ppf(alpha / 2, dfn, dfd)
f_crit_high = f.ppf(1 - alpha / 2, dfn, dfd)

# Step 5: Print results
print("🔍 F-test for Equality of Variances:")
print(f"Sample 1 Variance = {var1:.2f}")
print(f"Sample 2 Variance = {var2:.2f}")
print(f"F-statistic       = {f_stat:.3f}")
print(f"F-critical values = [{f_crit_low:.3f}, {f_crit_high:.3f}]")

# Step 6: Interpretation
if f_stat < f_crit_low or f_stat > f_crit_high:
    print(" Conclusion: Reject H₀ → Variances are significantly different.")
else:
    print(" Conclusion: Fail to reject H₀ → No significant difference in variances.")

# Step 7: Visualization
x = np.linspace(0.1, 5, 500)
y = f.pdf(x, dfn, dfd)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='F-distribution PDF', color='blue')
plt.axvline(f_stat, color='red', linestyle='--', label=f'F-statistic = {f_stat:.2f}')
plt.axvline(f_crit_low, color='green', linestyle='--', label=f'Lower critical value = {f_crit_low:.2f}')
plt.axvline(f_crit_high, color='green', linestyle='--', label=f'Upper critical value = {f_crit_high:.2f}')
plt.fill_between(x, 0, y, where=(x < f_crit_low) | (x > f_crit_high), color='red', alpha=0.2, label='Rejection region')

plt.title("F-distribution and Test Statistic")
plt.xlabel("F-value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Step 1: Simulate observed data (e.g., dice outcomes)
np.random.seed(42)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=120, replace=True)
obs_counts = np.bincount(observed)[1:]  # Skip index 0

# Step 2: Define expected frequencies (fair die → equal probability)
expected_counts = np.array([20, 20, 20, 20, 20, 20])  # 120 total / 6 sides

# Step 3: Perform Chi-square goodness-of-fit test
chi2_stat, p_val = chisquare(f_obs=obs_counts, f_exp=expected_counts)

# Step 4: Output the results
print("🎲 Chi-Square Goodness-of-Fit Test:")
print(f"Observed Counts: {obs_counts}")
print(f"Expected Counts: {expected_counts}")
print(f"Chi-square Statistic = {chi2_stat:.3f}")
print(f"P-value              = {p_val:.4f}")

# Step 5: Interpret the result
alpha = 0.05
if p_val < alpha:
    print(" Conclusion: Reject H₀ → Observed distribution does NOT fit expected.")
else:
    print(" Conclusion: Fail to reject H₀ → No significant difference from expected.")

# Step 6: Visualization
labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, obs_counts, width, label='Observed', color='skyblue')
plt.bar(x + width/2, expected_counts, width, label='Expected', color='orange')
plt.xticks(x, labels)
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title('Chi-square Goodness-of-Fit: Observed vs Expected')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
